```python
import os
import json
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, BatchNormalization, ReLU, Softmax, Dropout, LeakyReLU
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.regularizers import l2
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt
```

```python
def get_maqam(file_name, labels_dir):
    # Remove '_clipX' from the filename to match the JSON filenames
    base_name = file_name.split('_clip')[0]
    json_filename = base_name + '.mp3.json'
    json_path = os.path.join(labels_dir, json_filename)
    with open(json_path, 'r') as f:
        data = json.load(f)
        return data['maqam']
```

```python
# CHROMA VERSION

def train_neural_network(csv_path, labels_dir, num_epochs=150, batch_size=64, learning_rate=0.0001, test_size=0.2):

    # Load the dataset
    df = pd.read_csv(csv_path)

    # Extract filenames and all features
    file_names = df.iloc[:, 0].values  # First column with filenames
    mfcc_columns = [col for col in df.columns if 'mfcc_mean' in col]
    chroma_columns = [col for col in df.columns if 'chroma' in col]
    rms_col = [col for col in df.columns if 'rms' in col]
    zcr_col = [col for col in df.columns if 'zcr' in col]
    centroid_col = [col for col in df.columns if 'spectral_centroid' in col]
    bandwidth_col = [col for col in df.columns if 'spectral_bandwidth' in col]
    spectral_col = [col for col in df.columns if 'spectral_rolloff' in col]

  
    
    #X = training_vals.values
    X = df[chroma_columns].values 
    #X = df[mfcc_columns].values

    # Get correct maqams for each file
    y = np.array([get_maqam(file_name, labels_dir) for file_name in file_names])

    # Encode the labels
    label_encoder = LabelEncoder()
    y_encoded = label_encoder.fit_transform(y)

    # Convert labels to categorical format
    y_categorical = to_categorical(y_encoded)

    # Split the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y_categorical, test_size=test_size, random_state=42)

    # Define the neural network architecture
    model = Sequential()

    # Layer 1: Input layer + first hidden layer
    model.add(Dense(2048, input_shape=(X_train.shape[1],)))
    model.add(Dropout(0.2))
    model.add(BatchNormalization())
    model.add(ReLU())

    # Layer 2: Second hidden layer
    model.add(Dense(1024))
    model.add(Dropout(0.2))
    model.add(BatchNormalization())
    model.add(ReLU())

    # Layer 3: Third hidden layer
    model.add(Dense(512))
    model.add(Dropout(0.1))
    model.add(BatchNormalization())
    model.add(ReLU())

    # Layer 4: Fourth hidden layer
    model.add(Dense(256))
    model.add(Dropout(0.1))
    model.add(BatchNormalization())
    model.add(ReLU())

    # Layer 5: Fifth hidden layer
    model.add(Dense(128))
    model.add(BatchNormalization())
    model.add(ReLU())

    # Layer 6: Sixth hidden layer
    model.add(Dense(64))
    model.add(BatchNormalization())
    model.add(ReLU())

    # Output layer
    model.add(Dense(8))  # Number of output classes (8 maqams)
    model.add(Softmax())  # Activation for classification

    # Compile the model
    model.compile(optimizer=Adam(learning_rate=learning_rate), loss='categorical_crossentropy', metrics=['accuracy'])

    # Print the model summary
    model.summary()

    # Train the model continuously
    history = model.fit(X_train, y_train, epochs=num_epochs, batch_size=batch_size, validation_split=0.2)
        
    # Evaluate the model on the test set after each iteration
    loss, accuracy = model.evaluate(X_test, y_test)
    print(f'Test Accuracy : {accuracy:.2f}')

    # Save the model after each iteration
    model.save(f'deep_ann_model_2025.h5')

    return model, history, label_encoder
```

```python
from sklearn.metrics import classification_report, confusion_matrix, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns

def evaluate_model(model, X_test, y_test, label_encoder):
    # Predict on the test data
    y_pred_probs = model.predict(X_test)  # Predicted probabilities
    y_pred = y_pred_probs.argmax(axis=1)  # Convert probabilities to class predictions
    y_true = y_test.argmax(axis=1)        # Convert one-hot labels to class indices

    # Compute overall metrics
    accuracy = (y_pred == y_true).mean()
    precision = precision_score(y_true, y_pred, average='weighted')
    recall = recall_score(y_true, y_pred, average='weighted')
    f1 = f1_score(y_true, y_pred, average='weighted')

    print(f"Overall Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")

    # Compute per-class metrics
    print("\nPer-Class Metrics:")
    class_names = label_encoder.classes_
    translated_class_names = ['Bayat', 'Hejaz', 'Rast', 'Seekah', 'Saba', 'Ajam', 'Kurd', 'Nahawand']
    
    
    print(classification_report(y_true, y_pred, target_names=translated_class_names))

    # Confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    print("\nConfusion Matrix:")
    print(cm)
    per_class_accuracy = cm.diagonal() / cm.sum(axis=1)  # TP / (TP + FN) for each class
    
    
    cm_normalized_row = 100*cm/cm.sum(axis=1, keepdims=True)
    cm_normalized_column = 100*cm / cm.sum(axis=0, keepdims=True)

    
    for idx, class_name in enumerate(translated_class_names):
        print(f"Accuracy for {class_name}: {per_class_accuracy[idx]:.4f}")

    # Visualize the confusion matrix

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=translated_class_names, yticklabels=translated_class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_normalized_row, annot=True, fmt='.1f', cmap='Blues', xticklabels=translated_class_names, yticklabels=translated_class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()
    
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_normalized_column, annot=True, fmt='.1f', cmap='Blues', xticklabels=translated_class_names, yticklabels=translated_class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()
    

    return per_class_accuracy, accuracy, precision, recall, f1, cm
```
